```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle
import os

# Load the newly expanded dataset
df = pd.read_csv('../data/Crop_recommendation.csv', on_bad_lines='skip')

print(f"Dataset loaded successfully with {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Separate the features (X) and the target/label (y)
X = df.drop('label', axis=1)
y = df['label']

# Initialize the Label Encoder
label_encoder = LabelEncoder()

# Transform the text labels into numbers
y_encoded = label_encoder.fit_transform(y)

# Let's look at the mapping to prove it worked
crop_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print(f"Total unique crops encoded: {len(crop_mapping)}")
```

```python
# Split the data (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Initialize the Random Forest Classifier
# n_estimators=100 means we are using 100 "trees" in our forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model! (This is where the actual learning happens)
print("Training the Random Forest model...")
rf_model.fit(X_train, y_train)
print("Training complete!")
```

```python
# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"🏆 Model Accuracy: {accuracy * 100:.2f}%")

# Generate a detailed classification report
report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)
print("\n--- Classification Report ---")
print(report)
```

```python
# Ensure the model directory exists
os.makedirs('../model', exist_ok=True)

# 1. Save the trained Random Forest model
pickle.dump(rf_model, open('../model/crop_model.pkl', 'wb'))

# 2. Save the Label Encoder (so the frontend can translate numbers back to text)
pickle.dump(label_encoder, open('../model/label_encoder.pkl', 'wb'))

# 3. Save the performance report as a text file for documentation
with open('../model/model_report.txt', 'w') as f:
    f.write(f"Model Accuracy: {accuracy * 100:.2f}%\n\n")
    f.write("Classification Report:\n")
    f.write(report)

print("✅ Model, Encoder, and Report successfully exported to the /model folder!")
```
